```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
import ast

def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):  # Replace 'text' with 'obj'
        if counter != 3:
            L.append(i['name'])
            counter += 1
    return L

```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics.pairwise import cosine_similarity


df = movies[['id', 'title', 'genres', 'cast', 'crew', 'keywords']].copy()

# Encodeing categorical features into numbers
label_encoders = {}
for col in ['genres', 'cast', 'crew', 'keywords']:
    label_encoders[col] = LabelEncoder()
    df[col] = label_encoders[col].fit_transform(df[col].astype(str))

# Normalizing movie_id
df['id'] = df['id'] / df['id'].max()

#  data to PyTorch tensors
X = torch.tensor(df[['id', 'genres', 'cast', 'crew', 'keywords']].values, dtype=torch.long)

#  PyTorch model
class MovieEmbeddingModel(nn.Module):
    def __init__(self, num_movies, num_genres, num_cast, num_crew, num_keywords, embedding_dim=50):
        super(MovieEmbeddingModel, self).__init__()
        self.movie_embed = nn.Embedding(num_movies, embedding_dim)
        self.genre_embed = nn.Embedding(num_genres, embedding_dim)
        self.cast_embed = nn.Embedding(num_cast, embedding_dim)
        self.crew_embed = nn.Embedding(num_crew, embedding_dim)
        self.keyword_embed = nn.Embedding(num_keywords, embedding_dim)

        self.fc = nn.Sequential(
            nn.Linear(embedding_dim * 5, 256),
            nn.ReLU(),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Linear(128, 1)  # Final embedding size
        )

    def forward(self, x):
        movie_vec = self.movie_embed(x[:, 0])
        genre_vec = self.genre_embed(x[:, 1])
        cast_vec = self.cast_embed(x[:, 2])
        crew_vec = self.crew_embed(x[:, 3])
        keyword_vec = self.keyword_embed(x[:, 4])

        concat_vec = torch.cat((movie_vec, genre_vec, cast_vec, crew_vec, keyword_vec), dim=1)
        return self.fc(concat_vec)

#  number of unique categories
num_movies = len(df['id'].unique())
num_genres = len(df['genres'].unique())
num_cast = len(df['cast'].unique())
num_crew = len(df['crew'].unique())
num_keywords = len(df['keywords'].unique())

# Initializing model
model = MovieEmbeddingModel(num_movies, num_genres, num_cast, num_crew, num_keywords)
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Train model
num_epochs = 10
for epoch in range(num_epochs):
    optimizer.zero_grad()
    outputs = model(X)
    loss = criterion(outputs, X[:, 0].float().view(-1, 1))  # Ensures correct shape

    loss.backward()
    optimizer.step()
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {loss.item()}")


movie_embeddings = model(X).detach().numpy()

# Computing similarity using Nearest neighbours
from sklearn.neighbors import NearestNeighbors

knn = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=10)
knn.fit(movie_embeddings)


def recommend(movie_name, df, movie_embeddings, knn):
    if movie_name not in df['title'].values:
        return "Movie not found in dataset."

    index = df[df['title'] == movie_name].index[0]  # Get the index of the movie
    distances, indices = knn.kneighbors([movie_embeddings[index]], n_neighbors=11)  # Use embeddings, not model

    recommendations = [df.iloc[i].title for i in indices[0][1:]]  # Exclude itself
    return recommendations



movie_name = "Inception"
recommendations = recommend(movie_name, df, movie_embeddings, knn)  # Pass embeddings, not model
print(recommendations)


```
